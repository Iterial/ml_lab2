Test and train data sets seam consistent: random instances are similar to the instances from the begining of the sets. Also, we can easily see only one (TARGET) column is absent in test data set, so it is our target variable. Let`s check in which columns and how much missing values we have.

Let`s remove all columns with more than 30% of missing value. Also, we shoud remove ID column, because it has no correlation with target, and separate target column form train set.

```python
target = train_set['TARGET'].astype('category')
train_set.drop('TARGET', axis='columns', inplace=True)

cols_to_drop = ['SK_ID_CURR']

for col in train_set:
    missing_values = train_set[col].isnull().sum() / train_set.shape[0]

    if missing_values > 0.3:
        cols_to_drop.append(col)

train_set.drop(columns=cols_to_drop, inplace=True)
test_set.drop(columns=cols_to_drop, inplace=True)

print('train set', train_set.shape)
print('test set', test_set.shape)
```

Before we deal with other missing values, let`s view which types of features our datasets have.

```python
for col in train_set:
    print(col, train_set[col].dtypes)
```

All object-type features are categorical, but also some of int64 features are. Let`s list all numerical features(because these are fewer) and mark all others as categorical.

```python
numerical_features = {"CNT_CHILDREN", "AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_ANNUITY",
                      "AMT_GOODS_PRICE", "REGION_POPULATION_RELATIVE", "DAYS_BIRTH",
                      "DAYS_EMPLOYED", "DAYS_REGISTRATION", "DAYS_ID_PUBLISH", "CNT_FAM_MEMBERS",
                      "HOUR_APPR_PROCESS_START", "EXT_SOURCE_2", "EXT_SOURCE_3", "OBS_30_CNT_SOCIAL_CIRCLE",
                      "DEF_30_CNT_SOCIAL_CIRCLE", "OBS_60_CNT_SOCIAL_CIRCLE", "DEF_60_CNT_SOCIAL_CIRCLE",
                      "DAYS_LAST_PHONE_CHANGE", "AMT_REQ_CREDIT_BUREAU_HOUR", "AMT_REQ_CREDIT_BUREAU_DAY",
                      "AMT_REQ_CREDIT_BUREAU_WEEK", "AMT_REQ_CREDIT_BUREAU_MON", "AMT_REQ_CREDIT_BUREAU_QRT",
                      "AMT_REQ_CREDIT_BUREAU_YEAR"}
categorical_features = []

for col in train_set:
    if not col in numerical_features:
        train_set[col] = train_set[col].astype("category")
        test_set[col] = test_set[col].astype("category")
        categorical_features.append(col)

for col in train_set:
    print(col, train_set[col].dtypes)
```

Now we can replace categorical missing values with 'XNA'. Numerical missing values will be replaced by respective columns means, when we will remove outliers.

```python
for col in categorical_features:
    if train_set[col].hasnans and not 'XNA' in train_set[col].cat.categories:
        train_set[col] = train_set[col].cat.add_categories('XNA')
    if test_set[col].hasnans and not 'XNA' in test_set[col].cat.categories:
        test_set[col] = test_set[col].cat.add_categories('XNA')

    train_set[col] = train_set[col].fillna('XNA')
    test_set[col] = test_set[col].fillna('XNA')
```

Before we proceed with checking datasets for duplicates and visualizing, we will check if there are categorical-type features from test data set, which have categories absent in train set.

```python
for col in categorical_features:
    unique_train = pd.unique(train_set[col])
    unique_test = pd.unique(test_set[col])
    for obj in unique_test:
        if not obj in unique_train:
            print(col, '|', unique_train, '|', unique_test, '|', obj)
```

As we see, -1 value is present in test data set, but isn`t in train set. Value of -1 for these feature seems to be a mark for missing value, so we we should replace it with the mean rating.

```python
mean = np.mean(train_set['REGION_RATING_CLIENT_W_CITY'].astype('int64')).round()
test_set['REGION_RATING_CLIENT_W_CITY'] = test_set['REGION_RATING_CLIENT_W_CITY'].replace(-1, mean)
```

As we can see, all documents exept 3, 6, 8 have close to no correlation eachbetween, and they all have substantialy lower corelation with TARGET then other features observed previously. We can keep FLAG_DOCUMENT_3, 6, 13, 16 and remove others.

Some AMT_REQ_CREDIT_BUREAU_# seem to have same problem (unbalanced), and they have lower correlation with TARGET then some other features, so we can remove AMT_REQ_CREDIT_BUREAU_HOUR and AMT_REQ_CREDIT_BUREAU_WEEK, as they have the lowes correlation value with TARGET among other numerical features. #_CNT_SOCIAL_CIRCLE also can be considered as unbalanced, but they have OK correlation values with TARGET, so we should not remove them. Looking at correlation matrix, we can remove some redundant features, which have high correlation value (>0.5 or <-0.5) with oter features. CNT_CHILDREN and CNT_FAM_MEMBERS have correlation value of 0.88, but CNT_CHILDREN has more correlation with TARGET, so we will remove CNT_FAM_MEMBERS. AMT_CREDIT, AMT_ANNUITY and AMT_GOODS_PRICE also have strong correlation eachbetween, so we will remove AMT_CREDIT and AMT_ANNUITY, considering AMT_GOODS_PRICE has the most correlation with TARGET among those three. DAYS_BIRTH and DAYS_EMPLOYED have strong negative correlation between each other, but they also have OK correlation with TARGET, so we will not remove any of those. And we again see #_CNT_SOCIAL_CIRCLE features, but now from perspective of correlation. Thinking likewise, we can freely remove OBS_60_CNT_SOCIAL_CIRCLE and DEF_60_CNT_SOCIAL_CIRCLE.

```python
to_remove = ['FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5', 'FLAG_DOCUMENT_7',
             'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9', 'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11',
             'FLAG_DOCUMENT_12', 'FLAG_DOCUMENT_14', 'FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_17',
             'FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21',
             'AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_WEEK', 'CNT_FAM_MEMBERS',
             'AMT_CREDIT', 'AMT_ANNUITY', 'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE']

for col in to_remove:
    train_set.drop(col, axis='columns', inplace=True)
    test_set.drop(col, axis='columns', inplace=True)

print(train_set.shape)
print(test_set.shape)
```

All this values in mensioned features looks like corrupted values, and we can simply remove then with nan and replace later. Although, outlier values of DAYS_EMPLOYED feature are very common, accounting for approximately 16% of values. Hovewer, we still need to remove them, or our model will be biased. And we can not just remove the entire feature, because it still contains very usefull data.

```python
train_set.loc[train_set['AMT_INCOME_TOTAL'] > 40000000, 'AMT_INCOME_TOTAL'] = np.nan
train_set.loc[train_set['AMT_REQ_CREDIT_BUREAU_QRT'] > 50, 'AMT_REQ_CREDIT_BUREAU_QRT'] = np.nan
train_set.loc[train_set['DAYS_EMPLOYED'] > 50000, 'DAYS_EMPLOYED'] = np.nan
test_set.loc[test_set['DAYS_EMPLOYED'] > 50000, 'DAYS_EMPLOYED'] = np.nan
train_set.loc[train_set['OBS_30_CNT_SOCIAL_CIRCLE'] > 50, 'OBS_30_CNT_SOCIAL_CIRCLE'] = np.nan
test_set.loc[test_set['OBS_30_CNT_SOCIAL_CIRCLE'] > 50, 'OBS_30_CNT_SOCIAL_CIRCLE'] = np.nan
```

Now, we should replace all nan in numerical features with some values. In general, it should be means of corresponding features, however, for some features, it should be 0 or some other specific value. For now, lets look how much nan values we have for each numeric feature in both sets.

```python
for col in numerical_features:
    print('train', col, train_set[col].isnull().sum())
    print('test', col, test_set[col].isnull().sum())
```

For each feature except AMT_REQ_CREDIT_BUREAU_# replacing nan values with means looks reasonable. For AMT_REQ_CREDIT_BUREAU_# we will replace each nan whith 0, because it looks like those values where omited while collecting data, if a previous one feature had a value (e.g.: if AMT_REQ_CREDIT_BUREAU_HOUR has value, than AMT_REQ_CREDIT_BUREAU_DAY hasn't).

```python
amt = ['AMT_REQ_CREDIT_BUREAU_DAY', 'AMT_REQ_CREDIT_BUREAU_MON',
       'AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_YEAR']
for col in amt:
    np.nan_to_num(train_set[col], copy=False, nan=0)
    np.nan_to_num(test_set[col], copy=False, nan=0)
```

```python
for col in numerical_features:
    mean = np.mean(train_set[col])
    np.nan_to_num(train_set[col], copy=False, nan=mean)
    np.nan_to_num(test_set[col], copy=False, nan=mean)
```

Last thing to do - move alredy labeled ordinal features from categorical to numerical, because we will need to perform one-hot encoding on other categorical features, which is not sutable for ordinal features.

```python
dif = ['REGION_RATING_CLIENT', 'REGION_RATING_CLIENT_W_CITY']
for col in dif:
    train_set[col] = train_set[col].astype('int64')
    test_set[col] = test_set[col].astype('int64')
    categorical_features = categorical_features.drop(col)

numerical_features = train_set.select_dtypes(exclude='category').columns
```

Now, when we processed our data sets, we can start modeling.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

We will be using logistic regression for our model. Before creating a model we will need to one-hot encode categorical features, so they can be processed by it, standartize our numerical features, so the model will be less biased, and split our train data into train and validation parts. (We can one-hot each categoric feature because each is nominal feature).

```python
data_transformer = ColumnTransformer(transformers=[
    ("categorical_features", OneHotEncoder(), categorical_features),
    ("numerical_features", StandardScaler(), numerical_features)
])
```

```python
train_features, validate_features, train_target, validate_target = train_test_split(train_set, target, test_size=0.05)

transformed_train_features = pd.DataFrame.sparse.from_spmatrix(data_transformer.fit_transform(train_features))
transformed_validate_features = pd.DataFrame.sparse.from_spmatrix(data_transformer.transform(validate_features))
```

```python
classification_model = LogisticRegression(class_weight='balanced', max_iter=2000)
classification_model.fit(transformed_train_features, train_target)
```

Now we can check our model on validation data. Because we have unbalanced data set, we also want to look at ROC_AUC score.

```python
train_preds = classification_model.predict(transformed_train_features)
validation_preds = classification_model.predict(transformed_validate_features)
```

```python
print("Accuracy on train set: ", accuracy_score(train_target, train_preds))
print("ROC_AUC on train set: ", roc_auc_score(train_target, train_preds))
print("Accuracy on validation set: ", accuracy_score(validate_target, validation_preds))
print("ROC_AUC on validation set: ", roc_auc_score(validate_target, validation_preds))
```

Finaly, we can predict on our test set.

```python
transformed_test_set = pd.DataFrame.sparse.from_spmatrix(data_transformer.transform(test_set))

test_preds = classification_model.predict(transformed_test_set)
```
